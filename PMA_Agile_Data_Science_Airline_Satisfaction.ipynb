```python
# ============================================================
# Question 1(a): Dataset Understanding
# Step 1: Import Required Libraries
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Display all columns
pd.set_option('display.max_columns', None)

print("Libraries imported successfully.")

# ============================================================
# Step 2: Mount Google Drive
# ============================================================

from google.colab import drive
drive.mount('/content/drive')

print("Google Drive mounted successfully.")

# ============================================================
# Step 3: Verify Dataset Location
# ============================================================

folder_path = "/content/drive/MyDrive"

print("Files in Google Drive:")
print(os.listdir(folder_path))
# ============================================================
# Step 4: Load Training Dataset
# ============================================================

train_df = pd.read_excel("/content/drive/MyDrive/train dataset.xlsx")

print("Training dataset loaded successfully.")

# Remove unnecessary index column
train_df = train_df.drop(columns=['Unnamed: 0'])

print("Unnecessary column 'Unnamed: 0' removed successfully.")
# ============================================================
# Step 5: Display Dataset Preview
# Required Evidence for Q1(a)
# ============================================================

print("="*70)
print("DATASET PREVIEW (FIRST FIVE RECORDS)")
print("="*70)

display(train_df.head())
# ============================================================
# Step 6: Display Dataset Information
# Required Evidence for Q1(a)
# ============================================================

print("="*70)
print("DATASET INFORMATION")
print("="*70)

train_df.info()

# ============================================================
# Step 7: Display Dataset Dimensions
# ============================================================

print("="*70)
print("DATASET DIMENSIONS")
print("="*70)

print(f"Number of Records   : {train_df.shape[0]}")
print(f"Number of Variables : {train_df.shape[1]}")

# ============================================================
# Step 8: Display Statistical Summary
# ============================================================

print("="*70)
print("STATISTICAL SUMMARY")
print("="*70)

display(train_df.describe())

# ============================================================
# Step 9: Display Variable Names
# ============================================================

print("="*70)
print("LIST OF VARIABLES")
print("="*70)

for column in train_df.columns:
    print(column)

# ============================================================
# Question 1(b): Exploratory Data Analysis (EDA)
# Step 10: Configure Visualization Style
# ============================================================

plt.style.use('default')
sns.set_theme(style="whitegrid")

print("Visualization style configured successfully.")

# ============================================================
# Step 11: Distribution Analysis
# Passenger Age Distribution
# ============================================================

plt.figure(figsize=(10,6))

sns.histplot(
    data=train_df,
    x='Age',
    bins=30,
    kde=True
)

plt.title("Distribution of Passenger Age")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")

plt.show()

# ============================================================
# Step 12: Relationship Analysis
# Correlation Heatmap
# ============================================================

plt.figure(figsize=(15,10))

numeric_columns = train_df.select_dtypes(include=['int64','float64'])

sns.heatmap(
    numeric_columns.corr(),
    cmap='coolwarm',
    annot=False
)

plt.title("Correlation Heatmap of Numerical Variables")

plt.show()

# ============================================================
# Step 13: Categorical Analysis
# Passenger Satisfaction by Travel Class
# ============================================================

plt.figure(figsize=(8,6))

sns.countplot(
    data=train_df,
    x='Class',
    hue='satisfaction'
)

plt.title("Passenger Satisfaction by Travel Class")
plt.xlabel("Travel Class")
plt.ylabel("Number of Passengers")

plt.legend(title="Satisfaction")

plt.show()

# ============================================================
# Question 1(c)(i) - Data Quality Check 1
# Check for Missing Values
# ============================================================

missing_values = train_df.isnull().sum()
missing_values = missing_values[missing_values > 0]

print("Missing Values:")
print(missing_values)
# ============================================================
# Question 1(c)(i) - Data Quality Check 2
# Check for Duplicate Records
# ============================================================

duplicate_records = train_df.duplicated().sum()

print("Duplicate Records:", duplicate_records)

# ============================================================
# Question 2(a) - Feature Engineering 1
# Label Encoding Categorical Variables
# ============================================================

from sklearn.preprocessing import LabelEncoder

# Create a copy of the dataset
df_fe = train_df.copy()

# Identify categorical columns
categorical_columns = df_fe.select_dtypes(include=['object']).columns

print("Categorical Columns:")
print(categorical_columns)

# Apply Label Encoding
le = LabelEncoder()

for column in categorical_columns:
    df_fe[column] = le.fit_transform(df_fe[column])

print("\nEncoded Dataset Preview:")
print(df_fe.head())

# ============================================================
# Question 2(a) - Feature Engineering 2
# Feature Scaling
# ============================================================

from sklearn.preprocessing import StandardScaler

# Numerical columns
numerical_columns = [
    'Age',
    'Flight Distance',
    'Departure Delay in Minutes',
    'Arrival Delay in Minutes'
]

# Fill missing values before scaling
df_fe['Arrival Delay in Minutes'] = df_fe['Arrival Delay in Minutes'].fillna(
    df_fe['Arrival Delay in Minutes'].median()
)

scaler = StandardScaler()

df_fe[numerical_columns] = scaler.fit_transform(df_fe[numerical_columns])

print("Scaled Numerical Features:")
print(df_fe[numerical_columns].head())

# Question 2(a) - Feature Engineering 3 (optional)
# Separate features and target variable

X = df_fe.drop('satisfaction', axis=1)
y = df_fe['satisfaction']

print("Feature Matrix Shape:", X.shape)
print("Target Variable Shape:", y.shape)

# ============================================================
# Question 2(b) - Baseline Model Development
# Random Forest Classifier
# ============================================================

# Import required libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# ------------------------------------------------------------
# Split the dataset into training and testing sets
# ------------------------------------------------------------

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("========================================")
print("Training and Testing Dataset")
print("========================================")
print("Training Set Shape :", X_train.shape)
print("Testing Set Shape  :", X_test.shape)

# ------------------------------------------------------------
# Develop the Baseline Model
# ------------------------------------------------------------

rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

# Train the model
rf_model.fit(X_train, y_train)

print("\nBaseline Model Successfully Trained!")

# ------------------------------------------------------------
# Make Predictions
# ------------------------------------------------------------

y_pred = rf_model.predict(X_test)

# ------------------------------------------------------------
# Model Evaluation
# ------------------------------------------------------------

accuracy = accuracy_score(y_test, y_pred)

print("\n========================================")
print("Baseline Model Evaluation")
print("========================================")
print(f"Accuracy Score : {accuracy:.4f}")

print("\nClassification Report")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, y_pred))

# ============================================================
# Question 2(c) - Improved Model Development
# XGBoost Classifier
# ============================================================

# Install XGBoost (Run once in Google Colab)
!pip install xgboost

# Import required library
from xgboost import XGBClassifier

# ------------------------------------------------------------
# Develop the Improved Model
# ------------------------------------------------------------

xgb_model = XGBClassifier(
    n_estimators=200,
    learning_rate=0.10,
    max_depth=6,
    random_state=42,
    eval_metric='logloss'
)

# Train the improved model
xgb_model.fit(X_train, y_train)

print("Improved Model Successfully Trained!")

# ------------------------------------------------------------
# Make Predictions
# ------------------------------------------------------------

y_pred_xgb = xgb_model.predict(X_test)

# ------------------------------------------------------------
# Model Evaluation
# ------------------------------------------------------------

accuracy_xgb = accuracy_score(y_test, y_pred_xgb)

print("\n========================================")
print("Improved Model Evaluation")
print("========================================")

print(f"Accuracy Score : {accuracy_xgb:.4f}")

print("\nClassification Report")
print(classification_report(y_test, y_pred_xgb))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, y_pred_xgb))

# ============================================================
# Question 3(a) - Automated Validation Script
# ============================================================

print("========================================")
print("Automated Data Validation")
print("========================================")

# ------------------------------------------------------------
# Check Missing Values
# ------------------------------------------------------------

print("\n1. Missing Values Check")

missing_values = df_fe.isnull().sum()
missing_columns = missing_values[missing_values > 0]

if len(missing_columns) == 0:
    print("PASS: No missing values found.")
else:
    print("WARNING: Missing values detected.")
    print(missing_columns)

# ------------------------------------------------------------
# Check Duplicate Records
# ------------------------------------------------------------

print("\n2. Duplicate Records Check")

duplicate_records = df_fe.duplicated().sum()

if duplicate_records == 0:
    print("PASS: No duplicate records found.")
else:
    print(f"WARNING: {duplicate_records} duplicate records found.")

print("\n========================================")
print("Data Validation Completed")
print("========================================")
```
